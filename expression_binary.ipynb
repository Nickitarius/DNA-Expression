```python
import numpy as np
import bisect

def read_input(filename):
    with open(filename, 'r') as f:
        lines = f.readlines()

    for i in range(len(lines)):
        lines[i] = lines[i].replace("\n", "")

    first_line = lines[0].split(" ")
    first_line = [int(numeric_string) for numeric_string in first_line]
    n = first_line[0]
    m = first_line[1]

    genes = []
    for i in range(1, 1+n):
        line = lines[i].split(" ")
        line = [int(numeric_string) for numeric_string in line]
        genes.append(line)

    reads = []
    for i in range(1+n, 1+m+n):
        line = lines[i].split(" ")
        line = [int(numeric_string) for numeric_string in line]
        reads.append(line)

    return n, m, genes, reads
```

```python
def binarySearchCustom(arr, x, moreLessEqual='equal'):
    low = 0
    high = len(arr) - 1
    mid = 0

    while low <= high:

        mid = (high + low) // 2

        # If x is greater, ignore left half
        if arr[mid] < x:
            low = mid + 1

        # If x is smaller, ignore right half
        elif arr[mid] > x:
            high = mid - 1

        # means x is present at mid
        else:
            return mid

    # print(f'{arr[mid]} {moreLessEqual} {x} mid {mid}')
    # If we reach here, then the element was not present
    if (arr[mid] > x and moreLessEqual == 'more'):
        # print('more\n')
        return mid
    elif ( moreLessEqual == 'less'): #arr[mid] < x and
        # print('less\n')
        return mid
    else:
        # print(-1)
        return -1
```
