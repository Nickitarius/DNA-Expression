```python
import numpy as np
import bisect


def read_input(filename):
    with open(filename, 'r') as f:
        lines = f.readlines()

    for i in range(len(lines)):
        lines[i] = lines[i].replace("\n", "")

    first_line = lines[0].split(" ")
    first_line = [int(numeric_string) for numeric_string in first_line]
    n = first_line[0]
    m = first_line[1]

    genes = []
    for i in range(1, 1+n):
        line = lines[i].split(" ")
        line = [int(numeric_string) for numeric_string in line]
        genes.append(line)

    reads = []
    for i in range(1+n, 1+m+n):
        line = lines[i].split(" ")
        line = [int(numeric_string) for numeric_string in line]
        reads.append(line)

    return n, m, genes, reads
```

```python
def binarySearchCustom(arr, x, xMoreLessEqual='equal'):
    low = 0
    high = len(arr) - 1
    mid = 0
    isXLower = True

    while low <= high:
        mid = (high + low) // 2

        # If x is greater, ignore left half
        if arr[mid] < x:
            low = mid + 1

        # If x is smaller, ignore right half
        elif arr[mid] > x:
            high = mid - 1

        # means x is present at mid
        else:
            # print(f'{arr[mid]} {xMoreLessEqual} than {x} mid {mid}')
            return mid

    if (arr[mid] > x):
        isXLower = True
    else:
        isXLower = False
        # if (mid-1 >= 0):

    print(
        f'Found x={x} must be {xMoreLessEqual} than arr[mid]={arr[mid]} mid {mid} isXLower {isXLower}')
    # If we reach here, then the element was not present
    if (isXLower and (xMoreLessEqual == 'less')):
        return mid
    if ((not isXLower) and (xMoreLessEqual == 'more')):
        return mid
    if (xMoreLessEqual == 'more' and isXLower ):#and mid>0
            return mid+1
    if (xMoreLessEqual == 'less' and (not isXLower) and mid<len(arr)-1):
        if (arr[mid+1]> x):
            return mid-1
    return -1
```

```python
# Для Stepik

# put your python code here
import numpy as np


def read_console_input():
    in1 = input()
    first_line = in1.split(" ")
    first_line = [int(numeric_string) for numeric_string in first_line]
    n = first_line[0]
    m = first_line[1]

    genes = []
    for i in range(1, 1+n):
        l = input()
        line = l.split(" ")
        line = [int(numeric_string) for numeric_string in line]
        genes.append(line)

    reads = []
    for i in range(1+n, 1+m+n):
        l = input()
        line = l.split(" ")
        line = [int(numeric_string) for numeric_string in line]
        reads.append(line)

    return n, m, genes, reads


def calculateExpressions():
    n, m, genes, reads = read_console_input()

    intersectionsReads = []
    intersectionsGenes = []
    # Iterate over genes
    for i in range(n):
        # Iterate over each gene interval
        for j in range(0, len(genes[i]), 2):
            # Iterate over reads
            for k in range(m):
                # Iterate over each read interval
                for r in range(0, len(reads[k]), 2):
                    isLeftIn = reads[k][r] >= genes[i][j] and \
                        reads[k][r] <= genes[i][j+1]
                    isRightIn = reads[k][r + 1] >= genes[i][j] and \
                        reads[k][r+1] <= genes[i][j+1]
                    isIncluding = reads[k][r] <= genes[i][j] and \
                        reads[k][r + 1] >= genes[i][j+1]
                    if (isLeftIn or isRightIn or isIncluding):
                        intersectionsGenes.append(i)
                        intersectionsReads.append(k)

    reads = []
    for g in range(n):
        temp = []
        for r in range(len(intersectionsGenes)):
            if (intersectionsGenes[r] == g):
                temp.append(intersectionsReads[r])

        reads.append(np.unique(temp).tolist())

    readsOccurencesCounts = np.zeros(m)
    for i in range(m):
        for g in reads:
            if (g.__contains__(i)):
                readsOccurencesCounts[i] = readsOccurencesCounts[i]+1

    for i in range(m):
        if (readsOccurencesCounts[i] > 1):
            for r in reads:
                if (r.__contains__(i)):
                    r.remove(i)

    uniqueReadsPerGene = np.zeros(n)
    for i in range(n):
        uniqueReadsPerGene[i] = len(reads[i])

    return uniqueReadsPerGene


reads = calculateExpressions()

for rd in reads:
    print(int(rd))
```
