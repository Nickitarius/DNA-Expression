```python
import numpy as np


def read_input(filename):
    with open(filename, 'r') as f:
        lines = f.readlines()

    for i in range(len(lines)):
        lines[i] = lines[i].replace("\n", "")

    first_line = lines[0].split(" ")
    n, m = [int(numeric_string) for numeric_string in first_line]
    # n = first_line[0]
    # m = first_line[1]

    genes = []
    for i in range(1, 1+n):
        line = lines[i].split(" ")
        line = [int(numeric_string) for numeric_string in line]
        genes.append(line)

    reads = []
    for i in range(1+n, 1+m+n):
        line = lines[i].split(" ")
        line = [int(numeric_string) for numeric_string in line]
        reads.append(line)

    return n, m, genes, reads
```

```python
def binarySearchCustom(arr, x, xMoreLessEqual='e'):
    low = 0
    high = len(arr) - 1
    mid = 0

    while low <= high:
        mid = (high + low) // 2

        # If x is greater, ignore left half
        if arr[mid] < x:
            low = mid + 1

        # If x is smaller, ignore right half
        elif arr[mid] > x:
            high = mid - 1

        # means x is present at mid
        else:
            return mid

    isXLower = x < arr[mid]
    
    if (not isXLower):
        if (xMoreLessEqual == 'l'):
            return mid
        elif (xMoreLessEqual == 'm'):
            if (mid % 2 == 0):
                return mid
            else:
                return mid+1
    elif (isXLower):
        if (xMoreLessEqual == 'm'):
            return mid
        elif (xMoreLessEqual == 'l'):
            if (mid % 2 != 0):
                return mid
            else:
                return mid-1

    return -1
```
