```python
import numpy as np


def read_input(filename):
    with open(filename, 'r') as f:
        lines = f.readlines()

    for i in range(len(lines)):
        lines[i] = lines[i].replace("\n", "")

    first_line = lines[0].split(" ")
    n, m = [int(numeric_string) for numeric_string in first_line]
    # n = first_line[0]
    # m = first_line[1]

    genes = []
    for i in range(1, 1+n):
        line = lines[i].split(" ")
        line = [int(numeric_string) for numeric_string in line]
        genes.append(line)

    reads = []
    for i in range(1+n, 1+m+n):
        line = lines[i].split(" ")
        line = [int(numeric_string) for numeric_string in line]
        reads.append(line)

    return n, m, genes, reads
```

```python
def binarySearchCustom(arr, x, xMoreLessEqual='e'):
    low = 0
    high = len(arr) - 1
    mid = 0

    while low <= high:
        mid = (high + low) // 2

        # If x is greater, ignore left half
        if arr[mid] < x:
            low = mid + 1

        # If x is smaller, ignore right half
        elif arr[mid] > x:
            high = mid - 1

        # means x is present at mid
        else:
            # print(f'{arr[mid]} {xMoreLessEqual} than {x} mid {mid}')
            return mid

    isXLower = arr[mid] > x
    # print(
    #     f'Found x={x} must be {xMoreLessEqual} than arr[mid]={arr[mid]} mid {mid} isXLower {isXLower}')
    # If we reach here, then the element was not present
    if (isXLower):
        if (xMoreLessEqual == 'l'):
            return mid
        elif (xMoreLessEqual == 'm'):
            return mid+1
    elif (not isXLower):
        if (xMoreLessEqual == 'm'):
            return mid
        elif (xMoreLessEqual == 'l'):
            return mid-1
            # if (not mid%2==1):
            #     return mid
            # else:
            #     return mid-1
    return -1
```

```python
def calculateExpressions(filename):
    n, m, genes, reads = read_input(filename)

    intersects = []

    for i in range(n):
        intersects.append([])

    # Iterate over reads
    for rd in range(m):
        # Iterate over each read interval
        for g in range(n):

            for i in range(0, len(reads[rd]), 2):
            # Iterate over genes
                L = binarySearchCustom(genes[g], reads[rd][i], 'm')
                R = binarySearchCustom(genes[g], reads[rd][i+1], 'l')
                isValid = R >= 0 and L >= 0
                isIntersect = (L <= R or R % 2 == 1 and L == R + 1)
                if isValid and isIntersect:
                    # print('yep\n')
                    intersects[g].append(rd)

        for j in range(rd+1):
            occurences = 0
            for gg in range(n):
                # print(f'g: {i} {g}')
                if intersects[gg].__contains__(j):
                    occurences += 1
                    if (occurences > 1):
                        for gene in range(n):
                            # print(f'gene: {i} {g} {gene}')
                            intersects[gene] = list(
                                filter(lambda a: a != j, intersects[gene]))

    uniqueReadsPerGene = []
    for i in range(n):
        uniqueReadsPerGene.append(len(np.unique(intersects[i])))

    return uniqueReadsPerGene
```
