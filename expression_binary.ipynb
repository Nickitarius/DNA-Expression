```python
import numpy as np
import bisect


def read_input(filename):
    with open(filename, 'r') as f:
        lines = f.readlines()

    for i in range(len(lines)):
        lines[i] = lines[i].replace("\n", "")

    first_line = lines[0].split(" ")
    first_line = [int(numeric_string) for numeric_string in first_line]
    n = first_line[0]
    m = first_line[1]

    genes = []
    for i in range(1, 1+n):
        line = lines[i].split(" ")
        line = [int(numeric_string) for numeric_string in line]
        genes.append(line)

    reads = []
    for i in range(1+n, 1+m+n):
        line = lines[i].split(" ")
        line = [int(numeric_string) for numeric_string in line]
        reads.append(line)

    return n, m, genes, reads
```

```python
def binarySearchCustom(arr, x, xMoreLessEqual='equal'):
    low = 0
    high = len(arr) - 1
    mid = 0
    isXLower = True

    while low <= high:
        mid = (high + low) // 2

        # If x is greater, ignore left half
        if arr[mid] < x:
            low = mid + 1

        # If x is smaller, ignore right half
        elif arr[mid] > x:
            high = mid - 1

        # means x is present at mid
        else:
            # print(f'{arr[mid]} {xMoreLessEqual} than {x} mid {mid}')
            return mid

    if (arr[mid] > x):
        isXLower = True
    else:
        isXLower = False
        # if (mid-1 >= 0):

    # print(
    #     f'Found x={x} must be {xMoreLessEqual} than arr[mid]={arr[mid]} mid {mid} isXLower {isXLower}')
    # If we reach here, then the element was not present
    if (isXLower and (xMoreLessEqual == 'less')):
        return mid
    if ((not isXLower) and (xMoreLessEqual == 'more')):
        return mid
    if (xMoreLessEqual == 'more' and isXLower):  # and mid>0
        return mid+1
    if (xMoreLessEqual == 'less' and (not isXLower) and mid < len(arr)-1):
        if (arr[mid+1] > x):
            return mid-1
    return -1
```

```python
def calculateExpressions(filename):
    n, m, genes, reads = read_input(filename)

    intersectionsReads = []
    intersectionsGenes = []

    for i in range(n):
        intersectionsReads.append([])

    for i in range(m):
        intersectionsGenes.append([])

    # Iterate over reads
    for rd in range(m):
        # Iterate over each read interval
        for i in range(0, len(reads[rd]), 2):
            # Iterate over genes
            for g in range(n):
                L = binarySearchCustom(genes[g], reads[rd][i], 'more')
                R = binarySearchCustom(genes[g], reads[rd][i+1], 'less')
                isValid = R >= 0 and L >= 0
                isIntersect = (L <= R or R % 2 == 1 and L == R + 1)
                if isValid and isIntersect:
                    # print('yep\n')
                    intersectionsReads[g].append(rd)
                    intersectionsGenes[rd].append(g)
            
        for j in range(rd+1):
            occurences = 0
            for gg in range(n):
                # print(f'g: {i} {g}')
                if intersectionsReads[gg].__contains__(j):
                    occurences += 1

                if (occurences > 1):
                    for gene in range(n):
                        # print(f'gene: {i} {g} {gene}')
                        intersectionsReads[gene] = list(
                            filter(lambda a: a != j, intersectionsReads[gene]))


    uniqueReadsPerGene = []
    for i in range(n):
        uniqueReadsPerGene.append(len(np.unique(intersectionsReads[i])))

    return uniqueReadsPerGene
```
